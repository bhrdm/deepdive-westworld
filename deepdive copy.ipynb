# Onderzoek informatie Delos

Data inladen.

```python
import json
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
import folium
from folium.plugins import HeatMap

data= pd.read_json('deep-dive-westworld-starter-kit/data/incident_reports.json')
data
```

## Incidenten type

Sorteren incindenten per type

```python
types = data['type'].unique()
types
```

Tellen van incidenten per types

```python
tellen=data['type'].value_counts()
print(tellen)
```

Cirkeldiagram van aantal incidenten per type

```python
plt.pie(tellen,labels=tellen.index , autopct=lambda p:f'{int(p * sum(tellen) / 100)}', startangle=140)
plt.legend(tellen.index, title = 'types', )
plt.title('cirkeldiagram van types')
plt.show()
```

Staafdiagram van aantal incidenten per type

```python
plt.figure(figsize=(4, 4))

tellen.plot(kind='bar', color='skyblue')

plt.xlabel('Type')
plt.ylabel('Aantal')

plt.title("Staafdiagram van Types (Aantal)")

plt.show()
```

## Tijdreekanalyse

Waneer zijn de incidenten gebeurt per dag van de week.

```python
data['timestamp'] = pd.to_datetime(data['timestamp'])
data['dagvanweek'] = data['timestamp'].dt.day_name()
tellingdag = data.groupby(['type', 'dagvanweek']).size().reset_index(name='aantal_gebeurtenissen')
plt.figure(figsize=(12, 6))

for incident_type in tellingdag['type'].unique():
    data1 = tellingdag[tellingdag['type'] == incident_type]
    plt.plot(data1['dagvanweek'], data1['aantal_gebeurtenissen'], marker='o', label=incident_type)
    
plt.xlabel('Dag van de week')
plt.ylabel('Aantal gebeurtenissen')

plt.title("Aantal gebeurtenissen per dag van de week per type")

plt.xticks(rotation=45)

plt.legend()
```

Waneer zijn de incidenten gebeurt per maand van het jaar.

```python
data['timestamp'] = pd.to_datetime(data['timestamp'])
data['dagvanmaand'] = data['timestamp'].dt.month_name()
tellingdag = data.groupby(['type','dagvanmaand']).size().reset_index(name='aantal_gebeurtenissen')
plt.figure(figsize=(12, 6))

for incident_type in tellingdag['type'].unique():
    data2 = tellingdag[tellingdag['type'] == incident_type]
    plt.plot(data2['dagvanmaand'], data2['aantal_gebeurtenissen'], marker='o', label=incident_type)
    
plt.xlabel('maand')
plt.ylabel('Aantal gebeurtenissen')

plt.title("Aantal gebeurtenissen per dag van de maand per type")

plt.xticks(rotation=45)

plt.legend()
```

## Heatmap

Hotspots tonen.

```python
import folium
from folium.plugins import HeatMap
import numpy as np

latitude_bins = 30
longitude_bins = 30

hist2d, xedges, yedges = np.histogram2d(
    data['longitude'], data['latitude'], bins=(longitude_bins, latitude_bins)
)

plt.figure(figsize=(10, 8))
sns.heatmap(hist2d.T, cmap='viridis', xticklabels=xedges, yticklabels=yedges)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Incident Heatmap')
plt.gca().invert_yaxis()

plt.show()
```

Hotspot op kaart brengen.

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

incident_counts = data.groupby(['latitude', 'longitude']).size().reset_index(name='incident_count')
background_image = mpimg.imread('deep-dive-westworld-starter-kit/images/map.png')

center_x = 1024
center_y = 1024

image_width = background_image.shape[1]
image_height = background_image.shape[0]

extent = [
    center_x - (2100 / 2),
    center_x + (2100/ 2),
    center_y - (2100 / 2),
    center_y + (2100/ 2)
]

fig, ax = plt.subplots()

ax.imshow(background_image, extent=extent)

heatmap = ax.scatter(incident_counts['longitude'], incident_counts['latitude'], c=incident_counts['incident_count'], cmap='hot', s=10)

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.colorbar(heatmap, label='Incident Count')

plt.show()
```

## Aantal bezoekers

Data inladen

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


dat= pd.read_json('deep-dive-westworld-starter-kit/data/visitor_data.json')
dat
```

Verander van datum naar jaar en maand.

```python
df = pd.DataFrame(dat)

df['Date'] = pd.to_datetime(df['Date'])

df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month

monthly_visitors = df.groupby(['Year', 'Month'])[['Total_Visitors', 'Male_Visitors', 'Female_Visitors']].sum().reset_index()

monthly_visitors
```

Aantal bezoekers per man en vrouw per datum.

```python
df = pd.DataFrame(dat)

df['Date'] = pd.to_datetime(df['Date'])

fig, ax = plt.subplots()

plt.figure(figsize=(10, 8))

ax.scatter(df['Date'], df['Male_Visitors'], label='Male Visitors', color='blue', alpha=0.5)

ax.scatter(df['Date'], df['Female_Visitors'], label='Female Visitors', color='red', alpha=0.5)

ax.scatter(df['Date'], df['Total_Visitors'], label='Total Visitors', color='green', alpha=0.5)

ax.set_xlabel('Date')
ax.set_ylabel('Total Visitors')
ax.set_title('Scatter Plot of Male vs Female Visitors with Total Visitors')

plt.xticks(df['Date'], rotation=50)

ax.legend()

fig.tight_layout()

plt.show()
```
